## Reading Image from File

Here we start by reading an image from a file

```python
import cv2
import matplotlib.pyplot as plt

print("Starting OpenCV Demo ...")
original_image = cv2.imread("./opencv-images/multiple_choice.jpg")
```

```python
def show(image, title="There Is No Title"):
  plt.axis('off')
  plt.title(title)
  plt.imshow(image)
```

```python
show(original_image)
```
